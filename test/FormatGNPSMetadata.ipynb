```python
import pandas as pd
```

```python
df = pd.read_csv("all_sampleinformation.tsv", sep="\t")
```

```python
#Renaming columns
rename_mapping = {}
for key in df.keys():
    if key == "filename":
        continue
    rename_mapping[key] = "Attribute_" + key
df = df.rename(columns=rename_mapping)
```

```python
#Removing filenames that are duplicated
import os
all_filenames = df["filename"].apply(lambda x: os.path.basename(x))
```

```python
filename_counts_df = all_filenames.value_counts()
```

```python
unique_filenames = [filename for filename in filename_counts_df.to_dict() if filename_counts_df[filename] == 1]
print(len(unique_filenames))
```

```python
#Filtering it out
df = df[df.apply(lambda x: os.path.basename(x["filename"]) in unique_filenames, axis=1)]
```
